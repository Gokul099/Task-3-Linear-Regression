There are some outliers in Newspaper for right side

how to find those outliers
*IQR - this method is used when data  is not normal
*3-Sigma rule  - use this method when data is normql

IQR

*FIND q1 and q3

*find IQR=q3-q1

*find upper limit and lowwer lmt

*lower lmt = Q1-1.5*IQR

UPPER LMT = Q3+1.5*IQR

find values which greater than upper lmt and lesser than lower lmt

```python
#find q1 and a3
Q1 = df['Newspaper'].quantile(0.25)
Q3 = df['Newspaper'].quantile(0.75)
print('lower quantile:',Q1)
print('upper quantial:',Q3)
```

```python
#FIND IQR
IQR = Q3-Q1
IQR
```

```python
#find lower lmt and upper lmt
lower_lmt = Q1-1.5*IQR
upper_lmt = Q3+1.5*IQR
print('lower_lmt:',lower_lmt)
print('upper_lmt:',upper_lmt)
```

```python
#find values above upper lmt
df.loc[df['Newspaper']>upper_lmt]
```

```python
#calculate percentage
len(df.loc[df['Newspaper']>upper_lmt])/len(df)
                                        
```

```python
df.loc[df['Newspaper']<lower_lmt]
```

how handle outliers

*data is not normal you can use median


```python
df.loc[df['Newspaper']>upper_lmt,'Newspaper']=df['Newspaper'].median()
```

```python
#split intom x Aand y
x = df.drop('Sales',axis = 1)
y = df['Sales']
```

split data for trainimg and testing



```python
#sklearn is a scentific libiray include ml package is cleanig and feature eng

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 4)
```

```python
x_train.shape
```

```python
x_test.shape
```

model buliding

```python
#import linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#train the model with x _ train , y - train
model.fit(x_train,y_train)
```

```python
#make prediction is test model make prediction
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

Evaluate the model

#loss function to know b/w actual and predicted values

different error/loss function in regression algorithm are,,

*MSE - mean square error

MAE = MEAN absolute error

RMSE - root mean square error

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
MSE = mean_squared_error(y_test,y_pred)
MSE
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MAE
```

```python
RMSE= np.sqrt(MSE)
RMSE
```

```python
R2 = r2_score(y_test,y_pred)
R2
```
